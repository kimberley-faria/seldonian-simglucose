```python
CR_ACTIONS = np.arange(3, 30, 2)
CF_ACTIONS = np.arange(5, 50, 2)

CR_ACTIONS.shape, CF_ACTIONS.shape
```

```python
cr = np.random.choice(CR_ACTIONS)
cf = np.random.choice(CF_ACTIONS)

cr, cf

ACTIONS = [*product(CR_ACTIONS, CF_ACTIONS)]
len(ACTIONS)
```

```python
import gym

# Register gym environment. By specifying kwargs,
# you are able to choose which patient to simulate.
# patient_name must be 'adolescent#001' to 'adolescent#010',
# or 'adult#001' to 'adult#010', or 'child#001' to 'child#010'
from gym.envs.registration import register
register(
    id='simglucose-adolescent2-v0',
    entry_point='simglucose.envs:ModT1DSimEnv',
    kwargs={'patient_name': 'adolescent#001'}
)

mod_env = gym.make('simglucose-adolescent2-v0')
```

```python
from simglucose.controller.bolus_controller import BolusController
from seldonian.dataset import Episode
from seldonian.utils.stats_utils import weighted_sum_gamma
```

```python
import time

def run_trial(env):
    total_reward = 0
    observation, reward, done, info = env.reset()
    a = np.random.choice(list(range(len(ACTIONS))))
    cr, cf = ACTIONS[a]
    controller = BolusController(cr=cr, cf=cf)
    
    t = 0
    start_time = time.time()
    while not done:
        action = controller.policy(observation, reward, done, patient_name='adolescent#001', meal=info['meal'])
        observation, reward, done, info = env.step(action)
        total_reward += reward
        t += 1
        if done:
            print(f"Episode finished after {t} (sample) timesteps, Simulation run time = {time.time() - start_time} secs.")
            break
            
    return t, a, cr, cf, total_reward
```

```python
episodes = []
```

```python

start = time.time()
for trial in range(1000):
    s_0 = 0
    t, a, cr_action, cf_action, reward = run_trial(mod_env)
    
    observations = [s_0]
    actions = [a]
    prob_actions = [1. / len(ACTIONS)]
    rewards = [reward]
    
    print(f"trial: {trial}", t, a, ACTIONS[a], cr_action, cf_action, reward)
    episodes.append(Episode(observations, actions, rewards, prob_actions))

print(f"Total generation time for {len(episodes)} episodes: {time.time() - start}")
```

```python
len(episodes)
```

```python
returns = np.array([weighted_sum_gamma(ep.rewards,gamma=1.0) for ep in episodes])
```

```python
J_random = np.mean(returns)
```
